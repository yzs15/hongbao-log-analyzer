```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def show(df):
    print(df[df.location == 1].describe())
    df[df.location == 1].exec.hist(bins=100, figsize=(15,4))
    plt.show()

    part_df = df[df.location == 2].copy()
    print(part_df.describe())
    tile = part_df.quantile(q=0.99).exec
    part_df[part_df.exec < tile].exec.hist(bins=100, figsize=(15,4))
    plt.show()
```

```python
df = pd.read_csv('/Volumes/Elements/exec-logs/0414013656-net-38400-50-5000-10mC20C60%-1/exec_time.csv')
df.exec = df.exec / 1000000
print(df.describe())
show(df)
```
