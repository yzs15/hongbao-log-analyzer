```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def show(df):
    print(df[df.location == 1].describe())
    df[df.location == 1].exec.hist(bins=100, figsize=(15,4))
    plt.show()

    part_df = df[df.location == 2].copy()
    print(part_df.describe())
    tile = part_df.quantile(q=0.99).exec
    part_df[part_df.exec < tile].exec.hist(bins=100, figsize=(15,4))
    plt.show()
```
