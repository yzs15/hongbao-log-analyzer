```python
fre = {}
for d in data:
    if d not in fre:
        fre[d] = 1
    else:
        fre[d] += 1

n = len(data)

entropy = 0
for ratio, count in fre.items():
    if count == 0:
        continue
    p = count / n
    pu = p*(1-ratio/100) / 2
    if pu != 0:
        entropy += -1 * pu * math.log2(pu)
print(entropy)
```
